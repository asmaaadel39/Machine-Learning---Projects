```python
import pandas as pd 
```

```python
df_fake=pd.read_csv("/kaggle/input/fake-and-true/Fake.csv")
df_fake.head()
```

```python
df_true=pd.read_csv("/kaggle/input/fake-and-true/True.csv")
df_true.head()
```

```python
df_fake["class"]=0
df_true["class"]=1
```

```python
df_fake.head()
```

```python
df_true.head()
```

**Merge two datasets together**

```python
data=pd.concat([df_fake,df_true],axis=0)
data
```

```python
data.columns
```

```python
df=data.drop(["title","subject","date"],axis=1)
df
```

**Random Shuffling dataframe**

```python
df=df.sample(frac=1)
df.head()
```

```python
df.reset_index(inplace=True)
df
```

```python
df.drop(["index"],axis=1,inplace=True)
df.head()
```

```python
import re
import string
```

```python
def wordopt(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub("\\W"," ",text) 
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)    
    return text
```

```python
df["text"]=df["text"].apply(wordopt)
```

```python
X=df["text"]
y=df["class"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(xv_train,y_train)
```

```python
pred=LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, pred))
```
