```python
df=pd.read_csv('/kaggle/input/datapr/data.csv',index_col='date',parse_dates=True)
```

```python
import matplotlib.pyplot as plt
df.plot(figsize=(12,8))
```

```python
start_date='2009-01-01'
end_date='2018-01-01'
mask=(df.index>=start_date) & (df.index<=end_date)
df=df.loc[mask]
```

```python
df.index.freq='MS'
df.plot(figsize=(12,8))
#now we plot the data for just 10 years
```

```python
dfyearly=df.rolling(window=12).mean()
```

```python
dfyearly.plot(title='Yearly AVG Gold Prices')
```
