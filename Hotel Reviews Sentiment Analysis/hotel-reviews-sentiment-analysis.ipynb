```python
def rating(rating):
    if rating>3 and rating<=5:
        return "Postive"
    if rating>0 and rating<=3:
        return "Negative"
df['Rating PosandNeg']=df['Rating'].apply(rating)
```

**Sentiment Analysis**

```python
import nltk
nltk.download("vader_lexicon")
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sentanalyze=SentimentIntensityAnalyzer()
```

```python
df['Scores']=df['Review'].apply(lambda review: sentanalyze.polarity_scores(review))
```

```python
df.head()
```

```python
df['compound']=df['Scores'].apply(lambda f:f["compound"])
```

```python
df.head()
```

```python
df['compound PosorNeg']=df['compound'].apply(lambda score: 'Positive'if score>=0 else 'Negative')
```

```python
df.head()
```

```python
Example='Hotel was nice and I enjoyed'
sentanalyze.polarity_scores(Example)
```

**Text Classification**

```python
X=df['Review']
y=df['Rating PosandNeg']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.pipeline import Pipeline 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
```

```python
textclf=Pipeline([('tfidf',TfidfVectorizer()),('clf',LinearSVC())])
```

```python
textclf.fit(X_train,y_train)
```

```python
pred=textclf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,pred))
```
